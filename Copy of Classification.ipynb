```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy
import random

df = pd.read_csv('wdbc.csv')
df_index = df.index
print(df)
```

```python
random_index = random.sample(list(df_index), 50)
df1 = df.loc[random_index]
print(df1)
```

```python
df1.dropna()
print(df1)
```

```python
Mdf = df1.loc[df['Diagnosis'] == 'M']
Bdf = df1.loc[df['Diagnosis'] == 'B']
print(Bdf)
```

```python
rMdf = Mdf.drop(["Diagnosis", "ID", "texture1", "perimeter1",  "smoothness1", "compactness1", "concavity1", "concave_points1"], axis='columns')
rBdf = Bdf.drop(["Diagnosis", "ID", "texture1", "perimeter1", "smoothness1", "compactness1", "concavity1", "concave_points1"], axis='columns')
print(rMdf)
```

```python
radius_Mdf = rMdf.loc[:, 'radius1'].mean()
radius_Bdf = rBdf.loc[:, 'radius1'].mean()
area_Mdf = rMdf.loc[:, 'area1'].mean()
area_Bdf = rBdf.loc[:, 'area1'].mean()

print(radius_Bdf)
```

```python
ypoints = numpy.random.uniform(rMdf.loc[:, 'radius1'])
plt.hist(ypoints)


ypoints = numpy.random.uniform(rBdf.loc[:, 'radius1'])
plt.hist(ypoints)

plt.show()
```
